__1.__ Реализовать оптимизацию методом градиентного спуска для x^2(np.sin(0,5_x)^2+1). Какие параметры шага и количества итераций оптимальны? При каких значениях шага оптимизиция не выходит из локальных минимумов?

Найдем производную:
### $(x^2\cdot(\sin^2(0,5\cdot x)+1))'=2x\cdot(\sin^2(0,5\cdot x)+1) + x^2 \cdot 2\cdot\sin(0.5\cdot x) \cos(0.5\cdot x) \cdot \frac{1}{2}=\\ 
= 2x\cdot(\sin^2(0,5\cdot x)+1) + x^2 \cdot \sin(0.5\cdot x) \cos(0.5\cdot x)$

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_1 = np.linspace(-50, 50, 100)
# Исходная функция
def funct(_x): return (_x**2)*(np.sin(0.5*_x)**2+1)
# Градиент функции - вектор указывающий направление наибольшего возрастания 
# функции (частная производная по _x, вычисленная в данной точке)
def gradient(_x): return (2*(_x)*(np.sin(0.5*_x)**2+1)+(_x**2)*(np.sin(0.5*_x)*np.cos(0.5*_x)))
```

```python
def calculate_gradient(func, x_1, delta=0.1): return (func(x_1+delta)-func(x_1))/delta
def calculate_gradient_2(func, x_1, delta=0.1): return (func(x_1+delta)-func(x_1-delta))/(2*delta)
```

```python
x_1=np.random.uniform(-50, 50)
print('true grad', gradient(x_1))
print('method1', calculate_gradient(funct, x_1))
print('method2', calculate_gradient_2(funct, x_1))
```

```python
class optimizer:
    def __init__(self, alpha, n_iter):
        self.alpha=alpha
        self.n_iter=n_iter
    def fit(self, func, grad, x0):
        for i in range(self.n_iter):
            x0-=self.alpha*grad(x0)
        return x0
opt1=optimizer(alpha, 100)
x=opt1.fit(funct, gradient, np.random.uniform(-50, 50))
print(x, funct(x))
```
